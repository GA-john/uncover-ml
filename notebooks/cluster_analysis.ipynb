```python
from os.path import splitext
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
cont = pd.read_csv('/home/sudipta/temp/uncover-ml/Test/cluster_contributions_final.csv', index_col='feature_names')
```

```python
for f in cont.columns:
    cov = cont[f]
    cov_mean = cov.values[1::3]
    cov_sd = cov.values[2::3]
    y_err = 3*cov_sd
    plt.figure()
    plt.errorbar(range(1, len(cov_mean) + 1) , cov_mean, yerr=y_err, 
            fmt='o', ecolor='g', capthick=2)
    plt.title(splitext(f)[0]) 
    
```
